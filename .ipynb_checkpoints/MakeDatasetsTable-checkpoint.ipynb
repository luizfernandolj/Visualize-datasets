```python
import os
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.preprocessing import LabelEncoder
```

```python
#path = "C:\\Users\\Luiz Fernando\\JupyterFiles\\Quantifier-project\\Datasets\\datasets"
path = "C:\\Users\\luiz_\\Jupyter\\Visualize-datasets\\datasets"

os.chdir(path)
```

```python
df = pd.DataFrame(columns=["name", "size", "Positives", "%Positive", "MaxSizeTest", "F1", "AUC"])
for dataset in os.listdir():
    d = pd.read_csv(f"{path}\\{dataset}")
    
    types = d.dtypes
    label_encoder = LabelEncoder()
    
    for i in range(len(types)):
        if(d.dtypes[i] == "object"):
            labels = label_encoder.fit_transform(d.iloc[:, i])
            d[d.columns[i]] = labels
    
    popColumn = d.pop("class")
    d = pd.concat([d, popColumn], axis=1)
    
    # Name
    name = dataset[:-4]
    # Size
    size = int(d.size)
    # Positives
    positives = dict(d["class"].value_counts())[1]
    # Positives
    positive = round((dict(d["class"].value_counts())[1]/size)*100)
    # MaxSizeTest
    maxSizeTest = d["class"].value_counts(ascending=True)[0]
    
    
    x = d.iloc[:, :-1]
    y = d.iloc[:, -1]
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=64)
    clf = RandomForestClassifier(n_estimators=200)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    
    # F1-score
    f1 = round(f1_score(y_test, y_pred), 2)
    # AUC
    auc = round(roc_auc_score(y_test, y_pred), 2)               
            
    arr = np.array([name, size, positives, positive, maxSizeTest, f1, auc])
    col = ["name", "size", "Positives", "%Positive", "MaxSizeTest", "F1", "AUC"]
    df = pd.concat([df, pd.DataFrame([arr], columns=col)], axis=0)

p = "C:\\Users\\luiz_\\Jupyter\\Visualize-datasets\\table.csv"
df.to_csv(p, index=False)
```
